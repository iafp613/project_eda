## Project purpose

It is well known by all that in recent years the life expectancy of the Spanish population, as well as the increase in the quality of care and the latest medical advances have led to a revolution in the improvement of hospitals, medical centers and treatments against various diseases .
On the contrary, the industrialization of a large number of artisanal processes, the abusive use of chemicals in food, the increase in environmental pollution, the mass exodus to cities and their stressful way of life, and globalization (which spreads diseases more rapidly , as we have recently verified with SARS-CoV-2); have led to an exponential growth of diseases that a few years ago were less common and an outbreak of new pathologies.

For this reason, I wanted to carry out an exploratory data analysis to verify whether the mortality rates in Spain have increased, or on the contrary, they have decreased.

With all the data obtained, I want to conclude whether in Spain, technology, medical advances and the lifestyle of the last forty years have contributed to the mortality increase or not.

## About the data

All data sets have been obtained from the website of the National Institute of Statistics of Spain [(INE)](https://www.ine.es/index.htm) and can be downloaded:
- Data set for deaths in Spain ---> [Deaths] (https://www.ine.es/jaxiT3/Tabla.htm?t=10803)
- Data set for the population in Spain ---> [Population] (https://www.ine.es/jaxiT3/Tabla.htm?t=10262)


The data obtained consists of 80 .csv files distributed as follows:
- A data set for the resident population in Spain each year between 1980 and 2019. Description:
    - It contains 5 columns (Sex, age, region, year and total population) and 19 rows (one for each region).
    
- A data set for each year with all the causes of death registered. Description:

    - It contains 6 columns (cause of death, sex, age, region, year and total deaths) and 2223 rows (one for each region and cause of death).

    - Some of death causes registered are:
    
        - Infectious and parasitic diseases.
        - Various types of tumors.
        - Blood and hematopoietic organs diseases.
        - Nervous system and sense organs diseases.
        - Chronic rheumatic heart diseases.
        - Atherosclerosis.
        - Respiratory system diseases.
        - Non-infectious enteritis and colitis.
        - Cirrhosis.
        - Musculoskeletal system and connective tissue diseases.
        - Genitourinary system diseases.
        - Deaths due to pregnancy, childbirth or the puerperium.
        - Congenital malformations, deformities and chromosomal abnormalities.
        - Symptoms, signs, and abnormal clinical and laboratory findings.
        - Traffic accidents.
        - Accidental poisoning by psychotropic drugs.
        - Suicide and Self-Injury.
        - Homicide.
        - Etc.

In total, I have obtained 537320 data, which I will have to sort, clean and adapt for the purpose of the study.

Find the Specific Death Rate

The Specific Death Rate is the proportion of people who die from a specific cause in a period in a population.

It is calculated as follows:

$$SDR = \frac{\text{Number of deaths from each cause}}{\text{Total population}}\cdot100$$

```python
df_completo_raw['TME'] = df_completo_raw.apply(
    lambda row: ((row.Total_defunciones / row.Total_población) * 100), axis=1)
```
